```python
text = "варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг" #записываем текст в переменную
s = text.split() #разбиваем текст на слова
bigrams = [] #создаем пустой список
for i in range(len(s) - 1): #проходимся циклом по всем словам (точнее по их индексам), чтобы определить, что такое биграмма
    bigram = "%s %s" % (s[i], s[i+1])
    print(bigram) #распечатываем каждую биграмму
    bigrams.append(bigram) #и добавляем ее в ранее созданный список
```

```python
poem = """
Удивительная кошка

Несчастная кошка порезала лапу-
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!
""" #сохраняем текст в переменную
print (len(poem)) #распечатываем количество символов в тексте
newpoem = [word.strip("!?.,-") for word in poem.split()] #генерируем список из слов, из которых вырезаем указанные знаки пунктуации
print (newpoem) #распечатываем его
print (len(newpoem)) #распечатываем количество объектов (слов) в этом списке
bigrams = [] #создаем пустой список для будущих биграмм
for i in range(len(newpoem) - 1): #проходимся циклом по всем словам (точнее по их индексам), чтобы определить, что такое биграмма
    bigram = "%s %s" % (newpoem[i], newpoem[i+1])
    print(bigram) #распечатываем каждую биграмму
    bigrams.append(bigram) #и добавляем ее в ранее созданный список
x = tuple(bigrams)
```
